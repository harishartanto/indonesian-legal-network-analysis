```python
import os
import json
import pandas as pd
```

```python
dir_path = 'data/json'
```

```python
all_data = []

for filename in os.listdir(dir_path):
    if filename.endswith('.json'):
        file_path = os.path.join(dir_path, filename)
        with open(file_path, 'r') as file:
            data = json.load(file)
            nama_file = filename
            nomor_peraturan = data.get('Nomor')
            judul = data.get('Judul')
            tahun = data.get('Tahun')
            nomor = data.get('No')
            
            first_konsideran_content = None
            first_article = None
            headers = []
            
            for block in data.get('Blocks', []):
                if block.get('Ref') == 'menimbang' and block.get('Type') == 'KONSIDERAN':
                    first_konsideran_content = block.get('Content', '')
                    break
                    
            for block in data.get('Blocks', []):
                if block.get('Type') == 'CONTENT_PASAL' and block.get('Pasal') == 'pasal-1':
                    first_article = block.get('Content', '')
                    break
                    
            for block in data.get('Blocks', []):
                # append block.get('Context', '') if not exist in headers
                if block.get('Context') not in headers and block.get('Context') != '' and block.get('Context') is not None:
                        headers.append(block.get('Context'))
                    
            if first_konsideran_content is not None:
                selected_data = {
                    'nama_file': nama_file,
                    'nomor_peraturan': nomor_peraturan,
                    'judul': judul,
                    'tahun': tahun,
                    'nomor': nomor,
                    'konsideran': first_konsideran_content,
                    'pasal_1': first_article,
                    'heading': headers
                }
                all_data.append(selected_data)

docs_df = pd.DataFrame(all_data)
docs_df
```

```python
# save to csv
docs_df.to_csv('data/law_analyzer_new99.csv', index=False)
```
